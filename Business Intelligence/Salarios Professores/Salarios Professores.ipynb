```python
import pandas as pd

dados = [
    ['MS', 13007.12, 'Centro-Oeste'],
    ['MA', 8452.03, 'Nordeste'],
    ['PA', 8289.86, 'Norte'],
    ['RR', 7700.47, 'Norte'],
    ['MT', 7343.44, 'Centro-Oeste'],
    ['PB', 6944.09, 'Nordeste'],
    ['RN', 6814.88, 'Nordeste'],
    ['AP', 6600.98, 'Norte'],
    ['DF', 6427.71, 'Centro-Oeste'],
    ['SE', 6176.76, 'Nordeste'],
    ['TO', 6130.70, 'Norte'],
    ['PE', 5841.91, 'Nordeste'],
    ['AL', 5767.15, 'Nordeste'],
    ['ES', 5685.97, 'Sudeste'],
    ['AM', 5631.18, 'Norte'],
    ['SP', 5565.00, 'Sudeste'],
    ['AC', 5370.35, 'Norte'],
    ['GO', 5160.49, 'Centro-Oeste'],
    ['RO', 5118.41, 'Norte'],
    ['RS', 5111.05, 'Sul'],
    ['SC', 5026.80, 'Sul'],
    ['PI', 4984.17, 'Nordeste'],
    ['BA', 4965.24, 'Nordeste'],
    ['CE', 4961.73, 'Nordeste'],
    ['PR', 4920.55, 'Sul'],
    ['MG', 4867.97, 'Sudeste']
]

df = pd.DataFrame(dados, columns=['UF', 'Salário inicial', 'Região'])

df.head()
```

```python
estado_menor = df.loc[df['Salário inicial'].idxmin()]
estado_maior = df.loc[df['Salário inicial'].idxmax()]

print(f"MENOR salário: {estado_menor['UF']} (R$ {estado_menor['Salário inicial']:,.2f})")
print(f"MAIOR salário: {estado_maior['UF']} (R$ {estado_maior['Salário inicial']:,.2f})")
```

```python
df_rank = df.sort_values(by='Salário inicial', ascending=False).reset_index(drop=True)

posicao_go = df_rank[df_rank['UF'] == 'GO'].index[0] + 1

print(f"O estado de Goiás (GO) está na {posicao_go}ª posição.")
```

```python
media_geral = df['Salário inicial'].mean()
print(f"Média salarial de TODOS os estados: R$ {media_geral:,.2f}")
```

```python
tabela_regioes = df.groupby('Região')['Salário inicial'].agg(
    Média='mean',
    Maior='max',
    Menor='min'
).round(2)

tabela_regioes = tabela_regioes.sort_values(by='Média', ascending=False)

tabela_regioes
```

```python
regiao_maior_media = tabela_regioes['Média'].idxmax()
regiao_menor_media = tabela_regioes['Média'].idxmin()

print(f"Região com MAIOR média: {regiao_maior_media}")
print(f"Região com MENOR média: {regiao_menor_media}")
```
